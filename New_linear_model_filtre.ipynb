```python
import pandas as pd
import numpy as np
import collections
import matplotlib.pyplot as plt
```

```python
train = pd.read_json ('train.json', lines=True)
```

```python
train.head(5)
```

```python
def prop_base(x):
    v = np.array(list(x))
    prop_A = np.mean(v == 'A')
    prop_C = np.mean(v == 'C')
    prop_U = np.mean(v == 'U')
    return prop_A, prop_C, prop_U
```

```python
def prop_loop_type(x):
    v = np.array(list(x))
    prop_S = np.mean(v == 'S')
    prop_E = np.mean(v == 'E')
    prop_H = np.mean(v == 'H')
    prop_B = np.mean(v == 'B')
    prop_I = np.mean(v == 'I')
    prop_X = np.mean(v == 'X')
    return prop_S,prop_E,prop_H,prop_B,prop_I,prop_X
```

```python
train_data = []
for mol_id in train['id'].unique():
    sample_data = train.loc[train['id'] == mol_id]
    pA, pC, pU = prop_base(sample_data['sequence'].values[0])
    pS, pE, pH, pB, pI, pX = prop_loop_type(sample_data['predicted_loop_type'].values[0])
    for i in range(68):
        sample_tuple = (sample_data['id'].values[0]+'_'+str(i), sample_data['sequence'].values[0][i],
                        pA,pC,pU,
                        sample_data['structure'].values[0][i], sample_data['predicted_loop_type'].values[0][i],
                        pS,pE,pH,pB,pI,pX,
                        sample_data['reactivity'].values[0][i], 
                        sample_data['deg_Mg_pH10'].values[0][i],
                        sample_data['deg_pH10'].values[0][i],
                        sample_data['deg_Mg_50C'].values[0][i],
                        sample_data['deg_50C'].values[0][i])
        train_data.append(sample_tuple)
    for i in range(68,107):
        sample_tuple = (sample_data['id'].values[0]+'_'+str(i), sample_data['sequence'].values[0][i],pA,pC,pU,
                        sample_data['structure'].values[0][i], sample_data['predicted_loop_type'].values[0][i],
                        pS,pE,pH,pB,pI,pX,
                        0, 
                        0,
                        0,
                        0,
                        0)
        train_data.append(sample_tuple)
```

```python
train_data = pd.DataFrame(train_data, columns=['id', 'sequence','prop_base_A','prop_base_C','prop_base_U', 'structure', 'predicted_loop_type',
                                               'prop_loop_S','prop_loop_E','prop_loop_H','prop_loop_B','prop_loop_I','prop_loop_X','reactivity', 'deg_Mg_pH10',
                                  'deg_pH10', 'deg_Mg_50C', 'deg_50C'])
```

```python
dummies_base = pd.get_dummies(train_data['sequence'])
dummies_loop = pd.get_dummies(train_data['predicted_loop_type'])
dummies_struct = pd.get_dummies(train_data['structure'])
```

```python
dummies_base = dummies_base.drop(columns = ['U'])
dummies_loop = dummies_loop.drop(columns = ['X'])
dummies_struct = dummies_struct.drop(columns = ['.'])
```
