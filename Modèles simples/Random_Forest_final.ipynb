## Ajout de donn√©es

```python
# Ajout des proportions des bases
def prop_base(x):
    v = np.array(list(x))
    prop_A = np.mean(v == 'A')
    prop_C = np.mean(v == 'C')
    prop_U = np.mean(v == 'U')
    return prop_A, prop_C, prop_U
```

```python
# Ajout des proportions des loop type
def prop_loop_type(x):
    v = np.array(list(x))
    prop_S = np.mean(v == 'S')
    prop_E = np.mean(v == 'E')
    prop_H = np.mean(v == 'H')
    prop_B = np.mean(v == 'B')
    prop_I = np.mean(v == 'I')
    prop_X = np.mean(v == 'X')
    return prop_S,prop_E,prop_H,prop_B,prop_I,prop_X
```

```python
# Ajout des distances des bases 
def get_dist(structure, seq_length):
    dist = np.zeros((seq_length))
    idx = []
    for i, token in enumerate(structure):
        if token == "(":
            idx.append(i)
        elif token == ")":
            j = idx.pop()  # index of the corresponding '(' 
            dist[i] = i-j
            dist[j] = i-j
    return (dist/seq_length)
```

```python
# Transformation des bases en format num√©rique
def transformation_base_nombre(x):
    if x == 'A':
        return 1
    if x == 'C':
        return 2
    if x == 'G':
        return 3
    if x == 'U':
        return 4
```

```python
# Ajout des nouvelles donn√©es dans notre base train
train_data = []
for mol_id in data['id'].unique():
    sample_data = data.loc[data['id'] == mol_id]
    pA, pC, pU = prop_base(sample_data['sequence'].values[0])
    pS, pE, pH, pB, pI, pX = prop_loop_type(sample_data['predicted_loop_type'].values[0])
    distance = get_dist(sample_data['structure'].values[0], sample_data['seq_length'].values[0])
    for i in range(sample_data['seq_scored'].values[0]):
        if i == 0:
            sample_tuple = (sample_data['id'].values[0]+'_'+str(i), transformation_base_nombre(sample_data['sequence'].values[0][i]),
                            0, transformation_base_nombre(sample_data['sequence'].values[0][i+1]),
                        pA,pC,pU,
                        sample_data['structure'].values[0][i], sample_data['predicted_loop_type'].values[0][i],
                        pS,pE,pH,pB,pI,pX,
                        distance[i],
                        i+1,    
                        sample_data['reactivity'].values[0][i], 
                        sample_data['deg_Mg_pH10'].values[0][i],
                        sample_data['deg_pH10'].values[0][i],
                        sample_data['deg_Mg_50C'].values[0][i],
                        sample_data['deg_50C'].values[0][i])
        else:
            sample_tuple = (sample_data['id'].values[0]+'_'+str(i), transformation_base_nombre(sample_data['sequence'].values[0][i]),
                            transformation_base_nombre(sample_data['sequence'].values[0][i-1]), transformation_base_nombre(sample_data['sequence'].values[0][i+1]),
                        pA,pC,pU,
                        sample_data['structure'].values[0][i], sample_data['predicted_loop_type'].values[0][i],
                        pS,pE,pH,pB,pI,pX,
                        distance[i],
                            i+1,
                        sample_data['reactivity'].values[0][i], 
                        sample_data['deg_Mg_pH10'].values[0][i],
                        sample_data['deg_pH10'].values[0][i],
                        sample_data['deg_Mg_50C'].values[0][i],
                        sample_data['deg_50C'].values[0][i])
        train_data.append(sample_tuple)
```

```python
# Conversion en dataframe
train_data = pd.DataFrame(train_data, columns=['id', 'base','base_prec','base_suiv','prop_base_A','prop_base_C','prop_base_U', 'structure', 'predicted_loop_type',
                                               'prop_loop_S','prop_loop_E','prop_loop_H','prop_loop_B','prop_loop_I','prop_loop_X',
                                               'distance_parite','position','reactivity', 'deg_Mg_pH10',
                                  'deg_pH10', 'deg_Mg_50C', 'deg_50C'])
```

```python
# Conversion des donn√©es cat√©gorielles en dummy
dummies_loop = pd.get_dummies(train_data['predicted_loop_type'])
dummies_struct = pd.get_dummies(train_data['structure'])
```

```python
# Suppression des colonnes qu'on utilise plus
dummies_loop = dummies_loop.drop(columns = ['X','S'])
dummies_struct = dummies_struct.drop(columns = ['.'])
```

```python
df = train_data.drop(["predicted_loop_type","structure"], axis = 1)
```

```python
# Dataframe final
df = pd.concat([df,dummies_loop,dummies_struct], axis = 1)
```

## Cr√©ation train et test

```python
nb_mol = X.shape[0]//107 # nb de mol√©cule diff√©rentes 
nb_mol
```

```python
liste_ind = np.arange(0,nb_mol) # Liste de 0 √† 2399 (soit 2400 √©l√©ments)
liste_ind
```

```python
# Cr√©√© une liste train compos√© de 80% des mol√©cules avec un ordre al√©atoire, Taille : 1920 
ind_train = np.random.choice(liste_ind,size = int(0.8*nb_mol), replace = False) # base de train 0.8/0.2 train/test
len(ind_train)
```

```python
# Cr√©ation du test avec les 20% des mol√©cules restantes, Taille : 480
ind_test = np.setdiff1d(liste_ind,ind_train)
len(ind_test)
```

```python
indice_train = np.array([]) 
for i in ind_train: # Pour chaque indice dans le train 
    liste_indice = np.arange(i*107,i*107+107) # Tableau allant de i*107 jusqu'√† i*107+107  (-1)
    indice_train = np.append(indice_train,liste_indice)
```

```python
indice_test = np.array([])
for i in ind_test:
    liste_indice = np.arange(i*107,i*107+107)
    indice_test = np.append(indice_test,liste_indice)
```

```python
X_train = X.loc[indice_train,:] # Taille (86349, 21)
Y_train = Y.loc[indice_train,:] # Taille (86349, 5)
```

```python
X_test = X.loc[indice_test,:] # (21614, 21)
Y_test = Y.loc[indice_test,:] # (21614, 5)
```
